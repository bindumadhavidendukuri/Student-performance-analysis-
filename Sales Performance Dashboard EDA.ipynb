## 2. Outlier Detection (IQR method) on Sales

```python
q1, q3 = df["Sales"].quantile([0.25, 0.75])
iqr = q3 - q1
lower, upper = q1 - 1.5 * iqr, q3 + 1.5 * iqr
outliers = df[(df["Sales"] < lower) | (df["Sales"] > upper)]
print(f"Sales outliers (IQR method): {len(outliers)} rows "
      f"({100*len(outliers)/len(df):.1f}% of data), bounds=({lower:.2f}, {upper:.2f})")
```
